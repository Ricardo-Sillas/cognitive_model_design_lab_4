## Part 1 of Model Creation

Part 1 of the model creation was to have the original plan that we were going with, which was having a 3x3 grid, and having the agents pick from this grid. From there if the agent was to land on a position with another agent, then the amount of points given to them would end up being 0. If the agent was to pick a position that is unique to only them, then they would be awarded the amount of points that the position has assigned to it. Part 1 of the model creation was used to print out the agents poitions picked to see if we could see any similarities while comparing them. For simplicity we currently have only agent one priting what they chose.

```python
import pyibl
import random
import matplotlib.pyplot as plt

from tqdm import tqdm

# Variables for the agents and model
PARTICIPANTS = 50
ROUNDS = 50
NOISE = 0.25
DECAY = 0.5
DEFAULT_UTILITY = 30

def run(rounds, participants):
    # Uses hashmaps with value of points for key and list of times value picked for its value
    choicesOne = {new_list: [0] * participants for new_list in range(1, 10)}
    choicesTwo = {new_list: [0] * participants for new_list in range(1, 10)}
    choicesThree = {new_list: [0] * participants for new_list in range(1, 10)}    
    choicesFour = {new_list: [0] * participants for new_list in range(1, 10)}
    # Number of times the same positions were selected by different agents
    sameChoice = [0] * participants

    # Creating all of the agents for the model
    aOne = pyibl.Agent(default_utility=DEFAULT_UTILITY, noise=NOISE, decay=DECAY)
    aTwo = pyibl.Agent(default_utility=DEFAULT_UTILITY, noise=NOISE, decay=DECAY)
    aThree = pyibl.Agent(default_utility=DEFAULT_UTILITY, noise=NOISE, decay=DECAY)
    aFour = pyibl.Agent(default_utility=DEFAULT_UTILITY, noise=NOISE, decay=DECAY)

    for p in range(participants):
        # Creating a grid for the agents to choose from
        grid = createGrid()
        for r in range(rounds):
            # Used to keep track of which positions were selected for every round
            same = []
            # Have the agents choose a position from the grid
            cOne = aOne.choose(*grid)
            cTwo = aTwo.choose(*grid)
            cThree = aThree.choose(*grid)
            cFour = aFour.choose(*grid)
            # Keeps track of the points that every agent ended up picking
            choicesOne[cOne][p] += 1
            choicesTwo[cTwo][p] += 1
            choicesThree[cThree][p] += 1
            choicesFour[cFour][p] += 1
            
            # Checks to see if any of the agents landed in any of the spots
            if cOne == cTwo or cOne == cThree or cOne == cFour:
                aOne.respond(0)
                same.append(cOne)
                sameChoice[p]+=1
            else:
                aOne.respond(cOne)
            if cTwo == cOne or cTwo == cThree or cTwo == cFour:
                aTwo.respond(0)
                if cTwo not in same:
                    same.append(cTwo)
                    sameChoice[p]+=1
            else:
                aTwo.respond(cTwo)
            if cThree == cOne or cThree == cTwo or cThree == cFour:
                aThree.respond(0)
                if cThree not in same:
                    same.append(cThree)
                    sameChoice[p]+=1
            else:
                aThree.respond(cThree)
            if cFour == cOne or cFour == cTwo or cFour == cThree:
                aFour.respond(0)
                if cFour not in same:
                    sameChoice[p]+=1
            else:
                aFour.respond(cFour)
        
        # Resets the agents
        aOne.reset()
        aTwo.reset()
        aThree.reset()
        aFour.reset()
        
    # Prints all of the values that the agents got for every iteration that they go through the model
    print(choicesOne)
#     print(choicesTwo)
#     print(choicesThree)
#     print(choicesFour)

# Used to create a 3x3 grid where every position is unique and has a number between 1 and 9
def createGrid():
    points = [*range(1,10)]
    grid = {}
    for i in range(9):
        value = random.choice(points)
        del points[points.index(value)]
        grid[value] = i
    return grid
    
def main():
    run(ROUNDS, PARTICIPANTS)

if __name__ == '__main__':
    main()
```

```python
import pyibl
import random
import matplotlib.pyplot as plt

from tqdm import tqdm

# Variables for the agents and model
PARTICIPANTS = 50
ROUNDS = 50
NOISE = 0.25
DECAY = 0.5
DEFAULT_UTILITY = 30

def run(rounds, participants):
    # Uses hashmaps with value of points for key and list of times value picked for its value
    choicesOne = {new_list: [0] * participants for new_list in range(1, 10)}
    choicesTwo = {new_list: [0] * participants for new_list in range(1, 10)}
    choicesThree = {new_list: [0] * participants for new_list in range(1, 10)}    
    choicesFour = {new_list: [0] * participants for new_list in range(1, 10)}
    # Number of times the same positions were selected by different agents
    sameChoice = [0] * participants
    
    # Creating all of the agents to pick the position of the model
    aOne = pyibl.Agent('Player 1', default_utility=DEFAULT_UTILITY, noise=NOISE, decay=DECAY)
    aTwo = pyibl.Agent('Player 2', default_utility=DEFAULT_UTILITY, noise=NOISE, decay=DECAY)
    aThree = pyibl.Agent('Player 3', default_utility=DEFAULT_UTILITY, noise=NOISE, decay=DECAY)
    aFour = pyibl.Agent('Player 4', default_utility=DEFAULT_UTILITY, noise=NOISE, decay=DECAY)
    
    # Creating all of the agents to determine to either stay or withdraw
    gridOne = pyibl.Agent('Select 1', ['stay', 'warning'], default_utility=DEFAULT_UTILITY, noise=NOISE, decay=DECAY)
    gridTwo = pyibl.Agent('Select 2', ['stay', 'warning'], default_utility=DEFAULT_UTILITY, noise=NOISE, decay=DECAY)
    gridThree = pyibl.Agent('Select 3', ['stay', 'warning'], default_utility=DEFAULT_UTILITY, noise=NOISE, decay=DECAY)
    gridFour = pyibl.Agent('Select 4', ['stay', 'warning'], default_utility=DEFAULT_UTILITY, noise=NOISE, decay=DECAY)

    for p in range(participants):
        # Creating a grid for the agents to choose from
        grid = createGrid()
        for r in range(rounds):
            # Used to keep track of which positions were selected for every round
            same = []
            # Have the agents choose one of the two positions that they chose
            cOne = aOne.choose(*grid)
            cTwo = aTwo.choose(*grid)
            cThree = aThree.choose(*grid)
            cFour = aFour.choose(*grid)
            # Keeps track of the points that every agent ended up picking
            choicesOne[cOne][p] += 1
            choicesTwo[cTwo][p] += 1
            choicesThree[cThree][p] += 1
            choicesFour[cFour][p] += 1
            
            # Warning for every choices made by the agents 
            warningOne = False
            warningTwo = False
            warningThree = False
            warningFour = False
            
            # Checks to see if any of the agents landed in any of the spots and gives the agents points accordingly
            if cOne == cTwo or cOne == cThree or cOne == cFour:
                warningOne = True
                stayingOne = gridOne.choose({'stay': True, 'warning': warningOne}, {'stay': False, 'warning': warningOne})['stay']
                if stayingOne:
                    aOne.respond(0)
                    gridOne.respond(0)
                else:
                    aOne.respond(int(cOne//2))
                    gridOne.respond(int(cOne//2))
                same.append(cOne)
                sameChoice[p]+=1
            else:
                aOne.respond(cOne)
                
            if cTwo == cOne or cTwo == cThree or cTwo == cFour:
                warningTwo = True
                stayingTwo = gridTwo.choose({'stay': True, 'warning': warningTwo}, {'stay': False, 'warning': warningTwo})['stay']
                if stayingTwo:
                    aTwo.respond(0)
                    gridTwo.respond(0)
                else:
                    aTwo.respond(int(cTwo//2))
                    gridTwo.respond(int(cTwo//2))
                if cTwo not in same:
                    same.append(cTwo)
                    sameChoice[p]+=1
            else:
                aTwo.respond(cTwo)
                
            if cThree == cOne or cThree == cTwo or cThree == cFour:
                warningThree = True
                stayingThree = gridThree.choose({'stay': True, 'warning': warningThree}, {'stay': False, 'warning': warningThree})['stay']
                if stayingThree:
                    aThree.respond(0)
                    gridThree.respond(0)
                else:
                    aThree.respond(int(cThree//2))
                    gridThree.respond(int(cThree//2))
                if cThree not in same:
                    same.append(cThree)
                    sameChoice[p]+=1
            else:
                aThree.respond(cThree)
                
            if cFour == cOne or cFour == cTwo or cFour == cThree:
                warningFour = True
                stayingFour = gridFour.choose({'stay': True, 'warning': warningFour}, {'stay': False, 'warning': warningFour})['stay']
                if stayingFour:
                    aFour.respond(0)
                    gridFour.respond(0)
                else:
                    aFour.respond(int(cFour//2))
                    gridFour.respond(int(cFour//2))
                if cFour not in same:
                    same.append(cFour)
                    sameChoice[p]+=1
            else:
                aFour.respond(cFour)
        
        # Checking the instances of all of the agents picking the positions
        aOne.instances()
#         aTwo.instances()
#         aThree.instances()
#         aFour.instances()
        # Checking the instances of all of the agents choosing to stay or withdraw
        gridOne.instances()
#         gridTwo.instances()
#         gridThree.instances()
#         gridFour.instances()
        
        # Reseting all of the agnets picking the positions
        aOne.reset()
        aTwo.reset()
        aThree.reset()
        aFour.reset()
        # Reseting all of the agents choosing to stay or withdraw
        gridOne.reset()
        gridTwo.reset()
        gridThree.reset()
        gridFour.reset()

# Used to create a 3x3 grid where every position is unique and has a number between 1 and 9
def createGrid():
    points = [*range(1,10)]
    grid = {}
    for i in range(9):
        value = random.choice(points)
        del points[points.index(value)]
        grid[value] = i
    return grid
    
def main():
    run(ROUNDS, PARTICIPANTS)

if __name__ == '__main__':
    main()
```

```python
import pyibl
import random
import matplotlib.pyplot as plt

from tqdm import tqdm

# Variables for the agents and model
PARTICIPANTS = 50
ROUNDS = 50
NOISE = 0.25
DECAY = 0.5
DEFAULT_UTILITY = 30

def run(rounds, participants):
    # Uses hashmaps with value of points for key and list of times value picked for its value
    choicesOne = {new_list: [0] * participants for new_list in range(1, 10)}
    choicesTwo = {new_list: [0] * participants for new_list in range(1, 10)}
    choicesThree = {new_list: [0] * participants for new_list in range(1, 10)}    
    choicesFour = {new_list: [0] * participants for new_list in range(1, 10)}
    # Number of times the same positions were selected by different agents
    sameChoice = [0] * participants
    
    # Creating all of the agents to pick the position of the model
    aOne = pyibl.Agent('Player 1', default_utility=DEFAULT_UTILITY, noise=NOISE, decay=DECAY)
    aTwo = pyibl.Agent('Player 2', default_utility=DEFAULT_UTILITY, noise=NOISE, decay=DECAY)
    aThree = pyibl.Agent('Player 3', default_utility=DEFAULT_UTILITY, noise=NOISE, decay=DECAY)
    aFour = pyibl.Agent('Player 4', default_utility=DEFAULT_UTILITY, noise=NOISE, decay=DECAY)
    
    # Creating all of the agents to determine to either stay or withdraw
    gridOne = pyibl.Agent('Select 1', ['stay', 'warning'], default_utility=DEFAULT_UTILITY, noise=NOISE, decay=DECAY)
    gridTwo = pyibl.Agent('Select 2', ['stay', 'warning'], default_utility=DEFAULT_UTILITY, noise=NOISE, decay=DECAY)
    gridThree = pyibl.Agent('Select 3', ['stay', 'warning'], default_utility=DEFAULT_UTILITY, noise=NOISE, decay=DECAY)
    gridFour = pyibl.Agent('Select 4', ['stay', 'warning'], default_utility=DEFAULT_UTILITY, noise=NOISE, decay=DECAY)

    for p in range(participants):
        # Creating a grid for the agents to choose from
        grid = createGrid()
        for r in range(rounds):
            # Used to keep track of which positions were selected for every round
            same = []
            # Have the agents choose one of the two positions that they chose
            cOne = aOne.choose(*grid)
            cTwo = aTwo.choose(*grid)
            cThree = aThree.choose(*grid)
            cFour = aFour.choose(*grid)
            # Keeps track of the points that every agent ended up picking
            choicesOne[cOne][p] += 1
            choicesTwo[cTwo][p] += 1
            choicesThree[cThree][p] += 1
            choicesFour[cFour][p] += 1
            
            # Warning for every choices made by the agents 
            warningOne = False
            warningTwo = False
            warningThree = False
            warningFour = False
            
            # Checks to see if any of the agents landed in any of the spots and gives agents points if they withdrew or chose a unique position
            if cOne == cTwo or cOne == cThree or cOne == cFour:
                warningOne = True
                stayingOne = gridOne.choose({'stay': True, 'warning': warningOne}, {'stay': False, 'warning': warningOne})['stay']
                if stayingOne:
                    saveAOne = aOne.respond()
                    saveGridOne = gridOne.respond()
                else:
                    aOne.respond(int(cOne//2))
                    gridOne.respond(int(cOne//2))
                    cOne = 0
                same.append(cOne)
                sameChoice[p]+=1
            else:
                aOne.respond(cOne)
                cOne = 0
                
            if cTwo == cOne or cTwo == cThree or cTwo == cFour:
                warningTwo = True
                stayingTwo = gridTwo.choose({'stay': True, 'warning': warningTwo}, {'stay': False, 'warning': warningTwo})['stay']
                if stayingTwo:
                    saveATwo = aTwo.respond()
                    saveGridTwo = gridTwo.respond()
                else:
                    aTwo.respond(int(cTwo//2))
                    gridTwo.respond(int(cTwo//2))
                    cTwo = 0
                if cTwo not in same:
                    same.append(cTwo)
                    sameChoice[p]+=1
            else:
                aTwo.respond(cTwo)
                cTwo = 0
                
            if cThree == cOne or cThree == cTwo or cThree == cFour:
                warningThree = True
                stayingThree = gridThree.choose({'stay': True, 'warning': warningThree}, {'stay': False, 'warning': warningThree})['stay']
                if stayingThree:
                    saveAThree = aThree.respond()
                    saveGridThree = gridThree.respond()
                else:
                    aThree.respond(int(cThree//2))
                    gridThree.respond(int(cThree//2))
                    cThree = 0
                if cThree not in same:
                    same.append(cThree)
                    sameChoice[p]+=1
            else:
                aThree.respond(cThree)
                cThree = 0
                
            if cFour == cOne or cFour == cTwo or cFour == cThree:
                warningFour = True
                stayingFour = gridFour.choose({'stay': True, 'warning': warningFour}, {'stay': False, 'warning': warningFour})['stay']
                if stayingFour:
                    saveAFour = aFour.respond()
                    saveGridFour = gridFour.respond()
                else:
                    aFour.respond(int(cFour//2))
                    gridFour.respond(int(cFour//2))
                    cFour = 0
                if cFour not in same:
                    same.append(cFour)
                    sameChoice[p]+=1
            else:
                aFour.respond(cFour)
                cFour = 0
            
            # Checks to see if any of the agents landed in any of the spots and updates the agents points accordingly
            if cOne != 0 and (cOne == cTwo or cOne == cThree or cOne == cFour):
                saveAOne.update(0)
                saveGridOne.update(0)
                same.append(cOne)
                sameChoice[p]+=1
            elif cOne != 0:
                saveAOne.update(cOne)
                saveGridOne.update(cOne)
            if cTwo != 0 and (cTwo == cOne or cTwo == cThree or cTwo == cFour):
                saveATwo.update(0)
                saveGridTwo.update(0)
                if cTwo not in same:
                    same.append(cTwo)
                    sameChoice[p]+=1
            elif cTwo != 0:
                saveATwo.update(cTwo)
                saveGridTwo.update(cTwo)
            if cThree != 0 and (cThree == cOne or cThree == cTwo or cThree == cFour):
                saveAThree.update(0)
                saveGridThree.update(0)
                if cThree not in same:
                    same.append(cThree)
                    sameChoice[p]+=1
            elif cThree != 0:
                saveAThree.update(cThree)
                saveGridThree.update(cThree)
            if cFour != 0 and (cFour == cOne or cFour == cTwo or cFour == cThree):
                saveAFour.update(0)
                saveGridFour.update(0)
                if cFour not in same:
                    sameChoice[p]+=1
            elif cFour != 0:
                saveAFour.update(cFour)
                saveGridFour.update(cFour)
        
        # Reseting all of the agnets picking the positions
        aOne.reset()
        aTwo.reset()
        aThree.reset()
        aFour.reset()
        # Reseting all of the agents choosing to stay or withdraw
        gridOne.reset()
        gridTwo.reset()
        gridThree.reset()
        gridFour.reset()
    
    # Creating a list of list showing all agents picking the highest points possible to them
    allChoices = [choicesOne[9],choicesTwo[9],choicesThree[9],choicesFour[9]]
    # Prints a graph for all agents individually
    for p in range(len(allChoices)):
        plt.plot(range(1, ROUNDS + 1), allChoices[p], label=f"Agent {p}")
        plt.ylabel("Choices for the 9 points")
        plt.xlabel("Rounds")
        plt.legend()
        plt.show()
    
    # Prints a graph for all agents together
    for p in range(len(allChoices)):
        plt.plot(range(1, ROUNDS + 1), allChoices[p], label=f"Agent {p}")
    plt.ylabel("Choices for the 9 points")
    plt.xlabel("Rounds")
    plt.legend()
    plt.show()
    for p in range(len(allChoices)):
        plt.plot(range(1, ROUNDS + 1), allChoices[p], label=f"Agent {p}")
    plt.ylabel("Choices for the 9 points")
    plt.xlabel("Rounds")
    plt.legend()
    plt.show()
    
    # Prints a graph showing how many time the agents chose the same position in every instance
    plt.plot(range(1, ROUNDS + 1), sameChoice, label="Same choices")
    plt.ylabel("Number of times agent chose same places")
    plt.xlabel("Rounds")
    plt.legend()
    plt.show()

# Used to create a 3x3 grid where every position is unique and has a number between 1 and 9
def createGrid():
    points = [*range(1,10)]
    grid = {}
    for i in range(9):
        value = random.choice(points)
        del points[points.index(value)]
        grid[value] = i
    return grid
    
def main():
    run(ROUNDS, PARTICIPANTS)

if __name__ == '__main__':
    main()
```
